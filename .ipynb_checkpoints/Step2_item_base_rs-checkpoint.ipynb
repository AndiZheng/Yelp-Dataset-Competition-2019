```python
def word_tokenize(x):
    import nltk
    return nltk.word_tokenize(x)
```

```python
def remove_punctuation(x):
    import re, string
    list_punct=list(string.punctuation)
    out = [ch for ch in x if ch not in list_punct]
    return out
```

```python
def remove_stop_words(x):
    from nltk.corpus import stopwords
    stop_words = stopwords.words('english')
    filteredWords = [w.lower() for w in x if w.isalpha() if w.lower() not in stop_words]
    return filteredWords
```

```python
def lemmatize(x):
    import nltk
    wnl = nltk.WordNetLemmatizer()
    lemmatize = [wnl.lemmatize(w) for w in x]
    return lemmatize 
```

```python
from pyspark.sql.functions import col
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
# 定义一个 udf 函数
# 返回类型为字符串类型
udf_word_tokenize = udf(word_tokenize)
udf_remove_punctuation = udf(remove_punctuation)
udf_remove_stop_words = udf(remove_stop_words)
udf_lemmatize = udf(lemmatize)
# 使用
p1 = review_raw_data.withColumn('processed_text',udf_word_tokenize(review_raw_data.text))
p2 = p1.withColumn('processed_text',udf_remove_punctuation(p1.processed_text))
p3 = p2.withColumn('processed_text',udf_remove_stop_words(p2.processed_text))
p4 = p3.withColumn('processed_text',udf_lemmatize(p3.processed_text))
```

```python
# Item-Item Similarity Matrix, I choose cosine_similarity here
from sklearn.metrics.pairwise import cosine_similarity
item_sim_mat = cosine_similarity(matrix_pd.T)
```

```python
item_sim_mat.shape
```

```python
item_sim_mat
```

```python
# Find Neighborhoos with highest similarity sort array 按行排序
least_to_most_sim_indexes = np.argsort(item_sim_mat, axis=1)
# Hyperparameter: number of neighborhoods
neighborhood_size = 10 # can tune neighborhood size , hyper parameter
neighborhoods = least_to_most_sim_indexes[:, -neighborhood_size:] ## least_to_most_sim_indexes[:, -neighborhood_size: -1]
```

```python
# pick a lucky user
from time import time
user_id = '-dErbI4sHSkRz6oxjy9L3g'

n_users = matrix_pd.shape[0]
n_items = matrix_pd.shape[1]

start_time = time()
# 返回非零值的index
items_rated_by_this_user = matrix_pd.loc[user_id,:].values.nonzero()[0]
# Just initializing so we have somewhere to put rating preds
out = np.zeros(n_items)
for item_to_rate in range(n_items):
    relevant_items_idx = np.intersect1d(neighborhoods[item_to_rate],
                                    items_rated_by_this_user,
                                    assume_unique=True)  # assume_unique speeds up intersection op
    relevant_items = matrix_pd.columns[relevant_items_idx].values
    print (relevant_items_idx)
    print (relevant_items)
    print (matrix_pd.loc[user_id, relevant_items].values)
    print (item_sim_mat[item_to_rate, relevant_items_idx])
    print (item_sim_mat[item_to_rate, relevant_items_idx].sum())
    out[item_to_rate] = matrix_pd.loc[user_id, relevant_items].values.dot(item_sim_mat[item_to_rate, relevant_items_idx].T) / item_sim_mat[item_to_rate, relevant_items_idx].sum()

pred_ratings = np.nan_to_num(out)
print(pred_ratings)
print("Execution time: %f seconds" % (time()-start_time))
```

```python
# Hyperparameter: Recommend n restaurant
n = 10

# Get item indexes sorted by predicted rating
item_index_sorted_by_pred_rating = list(np.argsort(pred_ratings))

# Find items that have been rated by user
item_index_rated_by_this_user = matrix_pd.loc[user_id,:].values.nonzero()[0]

# We want to exclude the items that have been rated by user
unrated_items_by_pred_rating = [item for item in item_index_sorted_by_pred_rating
                                if item not in item_index_rated_by_this_user]

matrix_pd.columns[unrated_items_by_pred_rating[-n:]].values
```
